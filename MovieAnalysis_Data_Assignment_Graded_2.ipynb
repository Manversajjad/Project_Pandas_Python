```python
import pandas as pd

# Assuming 'movies' is your DataFrame
# Sample threshold for top-rated movies (adjust if necessary)
top_rated_movies = movies[movies['IMDb_rating'] >= 8.0]  # Change threshold if needed

# Step 1: Count the number of genres for each movie
# Count non-null genre columns for each movie
top_rated_movies['genre_count'] = top_rated_movies[['genre_1', 'genre_2', 'genre_3']].notnull().sum(axis=1)

# Step 2: Calculate the percentage for each genre count
genre_counts = top_rated_movies['genre_count'].value_counts(normalize=True) * 100

# Extract percentages for movies with exactly 1, 2, and 3 genres
percent_1_genre = genre_counts.get(1, 0)  # percentage of movies with 1 genre
percent_2_genres = genre_counts.get(2, 0)  # percentage of movies with 2 genres
percent_3_genres = genre_counts.get(3, 0)  # percentage of movies with 3 genres

# Display results
print(f"Percentage of top-rated movies with 1 genre: {percent_1_genre:.2f}%")
print(f"Percentage of top-rated movies with 2 genres: {percent_2_genres:.2f}%")
print(f"Percentage of top-rated movies with 3 genres: {percent_3_genres:.2f}%")
```

# Convert the budget and gross values from “dollars” to “dollars in millions” for all top-rated movies. Round the converted values down to 3 decimal places. For example, a value of 192,345,273 should be converted to 192.345.

```python
# Comverting Budget into Millions
movies['budget'] = (movies['budget'] / 1000000).map("${:,.3f}M".format)
print(movies)
```

```python
# Comverting Gross into Millions
movies['Gross'] = (movies['Gross'] / 1000000).map("${:,.3f}M".format)
print(movies)
```

# List all details for the top 10 movies with the highest profit, sorted from highest to lowest. Hint: Profit is not a column in the DataFrame. You will need to calculate it.

```python
movies['profit'] = movies['Gross'] - movies['budget']

top_Ten_Profit = movies['profit'].sort_values(ascending=False)
top_Ten_Profit.head(10)
```

# **Create a data visualization that shows each country and the number of top-rated movies produced in it. Find the country that produced the most top-rated movies.**

```python
import matplotlib.pyplot as plt
```

```python
# Step 1: Filter top-rated movies (IMDb rating >= 8.2)
top_rated_movies = movies[movies['IMDb_rating'] >= 8.2]

# Step 2: Count the number of top-rated movies by country
country_counts = top_rated_movies['Country'].value_counts()

# Step 3: Plot the data
plt.figure(figsize=(10, 6))
country_counts.plot(kind='bar', color='skyblue')
plt.title("Number of Top-Rated Movies by Country")
plt.xlabel("Country")
plt.ylabel("Number of Top-Rated Movies")
plt.xticks(rotation=45)
plt.show()

# Step 4: Find the country that produced the most top-rated movies
top_country = country_counts.max()
top_country_count = country_counts.max()

print(f"The country that produced the most top-rated movies is {top_country} with {top_country_count} movies.")
```
